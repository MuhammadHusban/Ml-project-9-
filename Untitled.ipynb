## Importing Libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import ExtraTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,precision_score,recall_score
import seaborn as sns
```

## Reading data

## Visualization 

```python
df.hist(bins  = 50,figsize = (20,15))
```

```python
def plot_continuous_columns(dataframe):
    numeric_columns = dataframe.select_dtypes(include=['number']).columns.tolist()
    dataframe = dataframe[numeric_columns]
    
    for i in range(0,len(numeric_columns),2):
        if len(numeric_columns) > i+1:
            plt.figure(figsize=(10,6))
            plt.subplot(121)
            sns.distplot(dataframe[numeric_columns[i]], kde=False)
            plt.subplot(122)            
            sns.distplot(dataframe[numeric_columns[i+1]], kde=False)
            plt.tight_layout()
            plt.show()

        else:
            sns.distplot(dataframe[numeric_columns[i]], kde=False)
def plot_box_plots(dataframe):
    numeric_columns = dataframe.select_dtypes(include=['number']).columns.tolist()
    dataframe = dataframe[numeric_columns]
    
    for i in range(0,len(numeric_columns),2):
        if len(numeric_columns) > i+1:
            plt.figure(figsize=(10,6))
            plt.subplot(121)
            sns.boxplot(dataframe[numeric_columns[i]])
            plt.subplot(122)            
            sns.boxplot(dataframe[numeric_columns[i+1]])
            plt.tight_layout()
            plt.show()

        else:
            sns.boxplot(dataframe[numeric_columns[i]])

print("Histograms\n")
plot_continuous_columns(df)  

print("===="*30)
print('\nBox Plots\n')
plot_box_plots(df)
            
```

```python
x = df.drop('MEDV',axis = 1)
y = df.get('MEDV')
```

## Train Test splitting

```python
x_train,x_test,y_train,y_test = train_test_split(x,y ,test_size = 0.2 ,random_state=42)
```

# Evaluating Models

```python
models = []
models.append(('Lr',LinearRegression()))
models.append(('Dr',DecisionTreeRegressor()))
models.append(('Et',ExtraTreeRegressor()))
models.append(('Ad',AdaBoostRegressor()))

models
```

```python
names = []
results =[]
```

```python
for name,model in models:
    obj = model
    obj.fit(x_train,y_train)
    y_pred= obj.predict(x_test)
    results.append((y_test,y_pred)*100)
    print(results)
```

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
y_train_pred = cross_val_score(obj,x_train,y_train,cv = 3)
print(y_train_pred)
```
